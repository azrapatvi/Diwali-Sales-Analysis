```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df
```

```python
df.info()
```

##Data Cleaning

```python
df.drop(columns={'Status','unnamed1'},inplace=True)
```

```python
df.drop_duplicates()
```

```python
df["Amount"].fillna(df['Amount'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
import pandas as pd
import numpy as np

date_range = pd.date_range(start="2022-01-01", end="2022-12-31", freq="D")
df["Date"] = np.random.choice(date_range, size=len(df))
```

```python
df.drop(columns=['Cust_name','User_ID'],inplace=True)
```

```python
df
```

##Sales Over Time

```python
import matplotlib.pyplot as plt

sales_over_time=df.groupby("Date")["Amount"].sum()
plt.figure(figsize=(15,5))
sales_over_time.plot(kind='line')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(axis='y',linestyle='--')
plt.title('Sales Over Time')
plt.show()
```

Monthly Trends → Spending per month (Date column)

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Amount'].sum()
print(monthly_sales)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))
monthly_sales.plot(kind='line', marker='o', color='teal')

plt.xlabel("Month")
plt.ylabel("Total Amount Spent")
plt.title("Monthly Spending Trend")
plt.grid(axis='y', linestyle='--')
plt.xticks(rotation=45)
plt.show()
```

##Product Analysis

Top Product Categories → Which Product_Category generates most sales.

```python
df['Product_Category'].value_counts() #transactions belong to which product category below cothing ans apparel has count of 2655 it means in all dataset Clothing & Apparel are purchased 2655 times
```

```python
no_of_trns_by_prod=df['Product_Category'].value_counts().head(10)
plt.figure(figsize=(15,5))

no_of_trns_by_prod.plot(kind='bar',color='#DDA0DD',edgecolor='black')
plt.xlabel('Product Category')
plt.ylabel('Number of Transactions')
plt.grid(axis='y',linestyle='--')
plt.title('Number of Transactions by Product Category')
plt.show()
```

```python
df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False)
```

```python
products_based_on_amount = df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(5)
explode=(0.2,0,0,0,0)
colors = ['#87CEEB',  # Sky Blue
          '#4682B4',  # Steel Blue
          '#9370DB',  # Medium Purple
          '#FFB6C1',  # Light Pink
          '#DDA0DD']  # Plum (Purple shade)

plt.pie(products_based_on_amount, labels=products_based_on_amount.index,explode=explode,colors=colors, autopct='%1.1f%%')
plt.title('Products Based on Amount')
plt.show()
```

##Customer Demographics & Spending

```python
gender_amount=df.groupby('Gender')['Amount'].sum()
gender_amount
```

Spending by Gender

```python
gender_amount=df.groupby('Gender')['Amount'].sum()
plt.figure(figsize=(10,5))
gender_amount.plot(kind='bar',color='#DDA0DD',edgecolor='black')
plt.xlabel('Gender')
plt.ylabel('Amount')
plt.grid(axis='y',linestyle='--')
plt.title('Amount spent by Gender')
plt.show()
```

Spending by Age Group

```python
ages_spend_alot=df.groupby('Age')['Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(15,5))
ages_spend_alot.plot(kind='bar',color='teal',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Amount')
plt.grid(axis='y',linestyle='--')
plt.title('Amount spent by Age')
for i, value in enumerate(ages_spend_alot):
    plt.text(i, value/2, str(value), ha='center', va='center', rotation=90, color='white', fontsize=11)
plt.show()
```

```python
plt.figure(figsize=(10,5))
age=df["Age"]
plt.hist(age, bins=5, color='#DDA0DD', edgecolor='black')  # raw spending values
plt.xlabel('Amount Spent')
plt.ylabel('Number of Transactions')
plt.grid(axis='y', linestyle='--')
plt.title('Distribution of Amount Spent')

plt.show()
```

Spending by Marital Status

```python
plt.figure(figsize=(10,5))
df.groupby('Marital_Status')['Amount'].sum().plot(kind='bar',color=['#DDA0DD','skyblue'],edgecolor='black')
plt.xlabel('Marital Status')
plt.ylabel('Amount')
plt.grid(axis='y',linestyle='--')
plt.title('Total Spending by Marital Status')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by Gender and Marital Status and sum the amount
spending = df.groupby(['Gender', 'Marital_Status'])['Amount'].sum().unstack()

print(spending)
ax=spending.plot(kind='bar', figsize=(10,6), color=['skyblue', '#DDA0DD'],edgecolor='black')

plt.xlabel('Gender')
plt.ylabel('Total Amount Spent')
plt.title('Spending by Gender and Marital Status')
plt.grid(axis='y', linestyle='--')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width()/2, p.get_height()), ha='center', va='bottom', fontsize=10)
plt.show()
```

##Occupation Insights

Top Spending Occupations:Who spends more: IT, Govt, Healthcare, Construction, etc.

```python
amount_spent_by_occupation=df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(15,5))
amount_spent_by_occupation.plot(kind='bar',color='#DDA0DD',edgecolor='black')
plt.xlabel('Occupation')
plt.ylabel('Amount')
plt.grid(axis='y',linestyle='--')
for i,value in enumerate(amount_spent_by_occupation):
  formatted_value = f"{value/1000:.1f}K" #ex:14802.3K=14802.3×1000=14,802,300 That’s 14.8 million
  plt.text(i,value/2,formatted_value,ha='center',va='center',rotation=90,fontsize=10)
plt.title('Amount spent by Occupation')
```

Occupation vs Product Category → Which occupations prefer which categories.\

```python
import matplotlib.pyplot as plt

occ_prod = df.groupby(['Occupation', 'Product_Category'])['Amount'].sum().unstack().fillna(0)

plt.figure(figsize=(15,6))
occ_prod.plot(kind='bar', stacked=True, figsize=(12,6), colormap='Pastel2', edgecolor='black')
plt.xlabel('Occupation')
plt.ylabel('Total Amount Spent')
plt.title('Occupation vs Product Category Spending')
plt.grid(axis='y', linestyle='--')
plt.legend(title='Product Category', bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()
```

## **Geographical Analysis**


Spending by State


```python
top_sales_by_state = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(5)

# Sort in ascending order for least sales
least_sales_by_state = df.groupby('State')['Amount'].sum().sort_values(ascending=True).head(5)

plt.figure(figsize=(12,5))

# Top 5
plt.subplot(1,2,1)
top_sales_by_state.plot(kind='bar', color='#DA70D6', edgecolor='black')
plt.xlabel('State')
plt.ylabel('Amount')
plt.grid(axis='y', linestyle='--')
plt.title('Top 5 Sales by State')

# Bottom 5 (lowest first)
plt.subplot(1,2,2)
least_sales_by_state.plot(kind='bar', color='#DDA0DD', edgecolor='black')
plt.xlabel('State')
plt.ylabel('Amount')
plt.grid(axis='y', linestyle='--')
plt.title('Lowest 5 Sales by State')

plt.tight_layout()
plt.show()
```

Spending by Zone

```python
top_sales_by_zone = df.groupby('Zone')['Amount'].sum().sort_values(ascending=False).head(5)

# Sort in ascending order for least sales
least_sales_by_zone = df.groupby('Zone')['Amount'].sum().sort_values(ascending=True).head(5)

plt.figure(figsize=(12,5))

# Top 5
plt.subplot(1,2,1)
top_sales_by_state.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Zone')
plt.ylabel('Amount')
plt.grid(axis='y', linestyle='--')
plt.title('Top 5 Sales by Zone')

# Bottom 5 (lowest first)
plt.subplot(1,2,2)
least_sales_by_zone.plot(kind='bar', color='#B0C4DE', edgecolor='black')
plt.xlabel('Zone')
plt.ylabel('Amount')
plt.grid(axis='y', linestyle='--')
plt.title('Lowest 5 Sales by Zone')

plt.tight_layout()
plt.show()
```

##Conclusion

- Customers with more orders generally spend more, showing a positive link between purchase frequency and spending.  
- Female customers spend more than males, and singles in both genders spend more than married customers.  
- Uttar Pradesh, Maharashtra, Karnataka, Delhi, and Madhya Pradesh lead in sales, while Telangana, Punjab, Rajasthan, Uttarakhand, and Jharkhand show the lowest sales.  
- North and West zones dominate sales, whereas the Eastern zone contributes the least; the Central zone performs moderately.  
- The IT sector spends the most, highlighting its strong purchasing power and demand for products.  
- Overall, high-order customers, female singles, and the IT sector drive most revenue, while weaker states and zones show opportunities for growth.
